```python
import pandas as pd
```

```python
data=pd.read_csv('insurance.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe(include='all')
```

```python
data['sex'].unique()
```

```python
data['sex']=data['sex'].map({'female':0,'male':1})
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].map({'yes':0,'no':1})
```

```python
data['region']=data['region'].map({'southwest':0, 'southeast':1, 'northwest':2, 'northeast':3})
```

```python
data.head()
```

```python
X = data.drop(['expenses'],axis=1)
```

```python
X
```

```python
Y= data['expenses']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr= LinearRegression()
lr.fit(X_train,Y_train)
```

```python
svm= SVR()
svm.fit(X_train,Y_train)
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train,Y_train)
```

```python
gbr=  GradientBoostingRegressor()
gbr.fit(X_train,Y_train)
```

```python
y_pred1=lr.predict(X_test)
y_pred2=svm.predict(X_test)
y_pred3=rfr.predict(X_test)
y_pred4=gbr.predict(X_test)
```

```python
print(y_pred1,y_pred2,y_pred3,y_pred4)
```

```python
df1=pd.DataFrame({'Actual':Y_test,'Linearregresion':y_pred1,'svm':y_pred2,'rfr':y_pred3,'gbr':y_pred4})
df1
```

```python
from sklearn import metrics
```

```python
score1= metrics.r2_score(Y_test,y_pred1)
score2= metrics.r2_score(Y_test,y_pred2)
score3= metrics.r2_score(Y_test,y_pred3)
score4= metrics.r2_score(Y_test,y_pred4)
print(score1,score2,score3,score4)
```

```python
s1= metrics.mean_absolute_error(Y_test,y_pred1)
s2= metrics.mean_absolute_error(Y_test,y_pred2)
s3= metrics.mean_absolute_error(Y_test,y_pred3)
s4= metrics.mean_absolute_error(Y_test,y_pred4)
print(s1,s2,s3,s4)
```
